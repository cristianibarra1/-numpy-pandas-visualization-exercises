```python
%matplotlib inline
import pandas as pd
import seaborn as sns
from pydataset import data
import matplotlib.pyplot as plt
from matplotlib import cm
import env
import os
import numpy as np
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
# What does the distribution of petal lengths look like?
sns.distplot(iris.petal_length)
```

```python
# Is there a correlation between petal length and petal width? Use http://guessthecorrelation.com/
# as a hint to how we could visually determine if a correlation exists between two numeric columns.
petal_corr = iris[['petal_length', 'petal_width']].corr()
sns.heatmap(petal_corr, annot=True)
```

```python
# Would it be reasonable to predict species based on sepal width and sepal length? For this, 
# you'll visualize two numeric columns through the lense of a categorical column.
plt.figure(figsize=(22 , 10))
sns.relplot(x='sepal_length', y='sepal_width', data=iris)
plt.title('Sepal size')

```

```python
# Which features would be best used to predict species?
sns.relplot(x='petal_length',y='petal_width', data=iris,hue='species')
plt.title('Species')
```

```python
def get_db_url(db, env_file=os.path.exists('env.py')):
    if env_file:
        user, password, host = (env.user, env.password, env.host)
        return f'mysql+pymysql://{user}:{password}@{host}/{db}'
    else:
        return False
url = get_db_url("chipotle")
```
